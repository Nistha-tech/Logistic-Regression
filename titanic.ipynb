```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
```

```python
train_data=pd.read_csv(r"C:\Users\hp\Downloads\0000000000002429_training_titanic_x_y_train.csv")
test_data=pd.read_csv(r"C:\Users\hp\Downloads\0000000000002429_test_titanic_x_test.csv")
```

```python
clf=LogisticRegression()
```

```python
df=train_data.copy()
df.shape
```

```python
df.head(5)
```

```python
df.shape
df.isnull().sum()
```

```python
def f(s):
    if s=="male":
        return 0
    else:
        return 1
def f1(s):
    if s=="S":
        return 2
    elif s=='C':
        return 1
    else :
        return 0
def f2(s):
    if s<=10:
        return 0
    elif s>10 and s<=20:
        return 1
    elif s>20 and s<=30:
        return 2
    else:
        return 3
def f3(s):
    if type(s)==float:
        return 0
    else:
        return 1
```

```python
df.Embarked.fillna('S',inplace=True)
df.Age.fillna(df.Age.mean(),inplace=True)
df['has_cabin']=df['Cabin'].apply(f3)
df['gender']=df['Sex'].apply(f)
df['name_len']=df['Name'].apply(len)
df['Embarked']=df['Embarked'].apply(f)
df['Fare']=df['Fare'].apply(f1)
df.drop(['Cabin','Sex','Name','Ticket'],axis=1,inplace=True)
df.head(5)
```

```python
test_data.Embarked.fillna('S',inplace=True)
test_data.Age.fillna(test_data.Age.mean(),inplace=True)
test_data['has_cabin']=test_data['Cabin'].apply(f3)
test_data['gender']=test_data['Sex'].apply(f)
test_data['name_len']=test_data['Name'].apply(len)
test_data['Embarked']=test_data['Embarked'].apply(f)
test_data['Fare']=test_data['Fare'].apply(f1)
test_data.drop(['Cabin','Sex','Name','Ticket'],axis=1,inplace=True)
test_data.head(5)
```

```python
x_train=df.drop("Survived",axis=1)
y_train=df["Survived"]
```

```python
print(x_train.shape)
print(y_train.shape)
print(test_data.shape)
```

```python
x_train.describe()
```

```python
clf.fit(x_train,y_train)
```

```python
test_data.Age.fillna(test_data.Age.mean(),inplace=True)
```

```python
y_pred=clf.predict(test_data)
y_pred.shape
```
